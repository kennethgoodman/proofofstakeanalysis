## Import libraries

```python
from numpy.random import choice
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Simulation Code

```python
def one_run(holdings, amount_gained):
    """
        Will increase the holdings of the winner (randomly chosen with respect to # of holdings) by 1
    """
    holdings_weight = holdings / holdings.sum()
    winner = choice(len(holdings_weight), p=holdings_weight)
    holdings[winner] += amount_gained
    return holdings
```

```python
def run_many(holdings, n, amount_gained):
    """
        return the holdings at the end of n trials and the proptions along the way
    """
    proportions = [holdings / holdings.sum()]
    for _ in range(n):
        holdings = one_run(holdings, amount_gained)
        proportions.append(holdings / holdings.sum())
    return holdings, proportions
```

```python
def run_and_plot_it(initial_holdings,X,trials=100000):
    Y = []
    for amount_gained in X:
        _, proportions = run_many(initial_holdings, trials, amount_gained)
        Y.append(np.max(np.array(proportions)[-1]))  # graph the participant with the most
    plt.plot(X,Y)
```

## Simulation With Two Parties

```python
initial_holdings = np.array([10,990])
X = np.linspace(.1,10,100)
run_and_plot_it(initial_holdings,X,trials=500000)
```
